## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

The Unit of Analysis: A single web page (url or page_id) per client domain, tracked weekly.

The Specific Decision: "Which subset of our underperforming or declining web pages should our content team prioritize for a rewrite or optimization check this week?"

The Action: A content writer or SEO analyst opens the generated dashboard, pulls up the top 20 flagged pages, reads the automated "reason codes," and manually audits the page content to update information, fix broken search intent, or add missing semantic details.

The Cost of a Wrong Call (The Risks):

False Positive (Type I Error): The model flags a perfectly healthy page as "declining." The Cost: An editor wastes 3–4 hours rewriting an article that didn't need it, incurring a direct labor cost and potentially breaking a page that was already performing fine.

False Negative (Type II Error): The model misses a page that is silently plummeting in organic impressions. The Cost: The client permanently loses their footprint on Google's Search Engine Results Pages (SERPs) to a competitor, leading to a long-term drop in conversion revenue and potential churn of the client account.
